**Data Cleaning**

```python
#Step3
df.isnull().sum()
```

```python
#Forward-fill to propagate the last valid value
df['CMPLNT_TO_DT'] = df['CMPLNT_TO_DT'].ffill()

#Backward-fill to propagate the next valid value
df['CMPLNT_TO_DT'] = df['CMPLNT_TO_DT'].bfill()

print(df['CMPLNT_TO_DT'].head())
```

**Feature Engineering**

Crime Count by Borough

```python
#Step4
#Group by borough and calculate total incidents
df['Total_Incidents'] = df.groupby('BORO_NM')['CMPLNT_NUM'].transform('count')
print(df['Total_Incidents'])
```

Crime Density Based on Latitude and Longitude

```python
# Round latitude and longitude to 2 decimal places to group by zones
df['Geo_Zone'] = df['Latitude'].round(2).astype(str) + "_" + df['Longitude'].round(2).astype(str)

# Calculate crime density for each zone
df['Crime_Density'] = df.groupby('Geo_Zone')['CMPLNT_NUM'].transform('count')
print(df['Crime_Density'])
```

Datetime Conversion

```python
#Ensure 'CMPLNT_FR_DT' is in datetime format
df['CMPLNT_FR_DT'] = pd.to_datetime(df['CMPLNT_FR_DT'])
print(df['CMPLNT_FR_DT'])
```

Weekend vs. Weekday

```python
# Step 1: Ensure 'CMPLNT_FR_DT' is in datetime format
df['CMPLNT_FR_DT'] = pd.to_datetime(df['CMPLNT_FR_DT'])

# Step 2: Create 'Day_Of_Week' column
df['Day_Of_Week'] = df['CMPLNT_FR_DT'].dt.day_name()
print("Feature 'Day_Of_Week' created.")

# Step 3: Define the function for weekend classification
def is_weekend(day):
    if day in ['Saturday', 'Sunday']:
        return 'Weekend'
    else:
        return 'Weekday'

# Step 4: Create 'Weekend_Indicator' column
df['Weekend_Indicator'] = df['Day_Of_Week'].apply(is_weekend)
print(df['Weekend_Indicator'])
```

**Data Visualization**

Bar Plot (Crime Count by Borough)


```python
#Step6
#Bar plot for total incidents by borough
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='BORO_NM', y='Total_Incidents')
plt.title('Total Crime Incidents by Borough')
plt.xlabel('Borough')
plt.ylabel('Total Incidents')
plt.show()
```

Heatmap (Crime Density by Time Category and Borough)


```python
#Create a pivot table for the heatmap
heatmap_data = pd.crosstab(df['BORO_NM'], df['Time_Category'])

#Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='coolwarm')
plt.title('Crime Density by Borough and Time Category')
plt.xlabel('Time Category')
plt.ylabel('Borough')
plt.show()
```

Crime Types across Boroughs

```python
#Group data by Borough and Offense Description and count incidents
heatmap_data = df.groupby(['BORO_NM', 'OFNS_DESC']).size().unstack(fill_value=0)

#Plot a heatmap using Seaborn
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='viridis', annot=False, linewidths=0.5)
plt.title('Crime Types Across Boroughs', fontsize=14)
plt.xlabel('Offense Type', fontsize=12)
plt.ylabel('Borough', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

Crime Count by Time Category


```python
#Pie chart for crime count by time category
time_counts = df['Time_Category'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(time_counts, labels=time_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Crime Distribution by Time Category')
plt.show()
```
